```python
def flatten(data,cols):
    for c in cols:
        fl = pd.DataFrame(dict(data[c])).T
        cols = [str(col) for col in fl.columns]
        fl.columns= [c+''+name for name in cols]
        data = pd.concat([data, fl],axis = 1)
        data = data.drop(c, axis = 1)
    return data
```

```python
from requests.exceptions import HTTPError

for url in ['https://api.github.com', 'https://api.github.com/invalid']:
    try:
        response = requests.get(url)
        # si nuestra conexion fue incorrecta lanzara un error
        response.raise_for_status()
    except HTTPError as http_err:
        print(f'Ocurrio un error {http_err}')
        
    except Exception as err:
        print(f'Ocurrio otro erro {err}')
    else:
        print('Todo salio Bien')
```

```python
def get_poke(n):
    res = requests.get(f'https://pokeapi.co/api/v2/pokemon/{n}')
    data = res.json()
    return {'name':data['name'], 'image':data['sprites']['front_default']}

```

```python
for i in range(9):
    print(get_poke(i+1))
```
