```python
data_phone[data_phone['item']== 'call'].groupby('month').agg(
            max_d = pd.NamedAgg(column = 'duration', aggfunc = max),
            min_d = pd.NamedAgg(column = 'duration', aggfunc = min),
            total = pd.NamedAgg(column = 'duration', aggfunc = sum))
            #num_days = pd.NamedAgg(column = 'date',  aggfunc = lambda x: (max(x)-min(x)).days))
```

```python
data_phone[data_phone['item']== 'call']
```

```python
aggregations = {
    # work on the "duration" column
    'duration': { 
        # get the sum, and call this result 'total_duration'
        'total_duration': 'sum',  
        # get mean, call result 'average_duration'
        'average_duration': 'mean', 
        'num_calls': 'count'
    },
    
     # Calculate two results for the 'network' column with a list
    'network': ["count", "max"] 
}
# Perform groupby aggregation by "month", 
# but only on the rows that are of type "call"
data_phone[data_phone['item'] == 'call'].groupby('month').agg(aggregations)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style = "white")
corr = carros.corr(method = 'spearman')
mask = np.triu(np.ones_like(corr, dtype = np.bool))
```

```python
mask
```

```python
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(150, 100, as_cmap=True)
sns.heatmap(corr, cmap=cmap, annot=True, mask = mask)
```
